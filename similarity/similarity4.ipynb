```python
import numpy as np
import skimage.color
import skimage.io
import matplotlib.pyplot as plt
import os
```

```python
# read the image of a plant seedling as grayscale from the outset
image_1 = skimage.io.imread(fname='images/nike1.jpeg', as_gray=True)
image_2 = skimage.io.imread(fname='images/nike2.png', as_gray=True)

# display the image
fig, ax = plt.subplots(nrows=1,ncols=2)
ax[0].imshow(image_1, cmap='gray')
ax[1].imshow(image_2, cmap='gray')
plt.show()
```

```python
# create the histogram
histogram_1, bin_edges_1 = np.histogram(image_1, bins=64, range=(0, 1))
histogram_2, bin_edges_2 = np.histogram(image_2, bins=64, range=(0, 1))
```

```python
# configure and draw the histogram figure
plt.figure(figsize=(20,5))
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixel count")
plt.xlim([0.0, 1.0])  # <- named arguments do not work here

plt.plot(bin_edges_1[0:-1], histogram_1)  # <- or here
plt.plot(bin_edges_2[0:-1], histogram_2)  # <- or here

plt.show()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error,r2_score

cosine_sim = cosine_similarity([histogram_1],[histogram_2])
rmse = mean_squared_error(histogram_1,histogram_2,squared=False)
r_squared = r2_score(histogram_1,histogram_2)
print(f"Cosine Similarity: {cosine_sim[0][0]}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-Squared: {r_squared}")
```

```python
def read_image(filename):
    return skimage.io.imread(fname=filename, as_gray=True)
```

```python
def check_histogram_similarity(image1,image2):
    histogram_1, bin_edges_1 = np.histogram(image1, bins=256, range=(0, 1))
    histogram_2, bin_edges_2 = np.histogram(image2, bins=256, range=(0, 1))
    return cosine_similarity([histogram_1],[histogram_2])
```

```python
def show_image(image_1,image_2):
    fig, ax = plt.subplots(nrows=1,ncols=2)
    ax[0].imshow(image_1, cmap='gray')
    ax[0].set_xlabel(filename_1[7:])
    ax[1].imshow(image_2, cmap='gray')
    ax[1].set_xlabel(filename_2[7:])
    plt.show()
```

```python
for filename_1 in filenames:
    for filename_2 in filenames:
        if not filename_1==filename_2:
            image_1 = read_image(filename_1)
            image_2 = read_image(filename_2)
            show_image(image_1,image_2)
            cosine_sim = check_histogram_similarity(image_1,image_2)
            print(f"Cosine Similarity: {cosine_sim[0][0]}")
            print()
```
