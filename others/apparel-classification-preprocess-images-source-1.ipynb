```python
import pandas as pd
import os
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
```

```python
data = pd.read_csv("data/image_labels.csv")
data
```

```python
apparel_list = ["T-Shirt","Longsleeve","Pants","Shirt","Dress","Shorts","Skirt"]
```

```python
from PIL import Image

image_list = []
for image in data.values:
    if image[2] in apparel_list:
        im1 = Image.open(f"data/apparel/{image[0]}.jpg")
        im1 = im1.save(f"data/apparels/{image[2]}/{image[0]}.jpg")
        if image[2] == "Polo":
            image[2] = "Shirt"
        image_list.append(image)
new_data = pd.DataFrame(data=image_list,columns=data.columns)
```

```python
new_data.drop(["sender_id","kids"],axis=1,inplace=True)
new_data.to_csv("data/image_labels.csv",index=False)
```
